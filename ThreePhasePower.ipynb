# Three Phase Power

Lets plot three sine waves corresponding to three line currents separated by 120 degrees.

Power to small appliances will typically use one of these three lines, whereas heavier duty factory induction motors will use all three at once for maximum power and efficiency.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create 500 points between zero and $5\pi$.

```python
domain = np.linspace(0, 5*np.pi, 500)
```

```python
def waveform(d, amplitude = 1, ϴ = 1, offset = 0):
    return amplitude * np.sin(ϴ * d + offset)
```

```python
ϕ_1 = waveform(domain)
ϕ_2 = waveform(domain, offset = np.radians(120))
ϕ_3 = waveform(domain, offset = np.radians(240))
```

```python
plt.title("Three Phase Current")
plt.xlabel("Radians")
plt.ylabel("Current")
plt.plot(domain, ϕ_1)
plt.plot(domain, ϕ_2)
plt.plot(domain, ϕ_3);
```
